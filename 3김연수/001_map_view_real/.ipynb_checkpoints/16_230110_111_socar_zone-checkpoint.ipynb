```python
#1. package load
import requests
import json
import os
import math
import time
import numpy as np
import pandas as pd
import geopandas as gpd
import pyproj
import folium
import matplotlib.pyplot as plt
# import cv2
import seaborn as sns
from IPython.display import Image

from shapely.geometry import mapping, shape, Point, Polygon, LineString
from folium import plugins
from folium.plugins import MarkerCluster, HeatMap

import warnings
warnings.filterwarnings(action='ignore') 

print(pd.__version__)
print(gpd.__version__)
print(folium.__version__)
```

```python
#2. 프로젝트(1)  ... data load
path = 'c:/users/user/000_my_jupyter/900_aiffel/0000_AIFFELTHON/3김연수/002_data_dir/004_node_data/20210823_socar_zone.csv'

socar_zone_info_df = pd.read_csv(path)
socar_zone_info_df.head()
```

```python
socar_zone_info_df[socar_zone_info_df['region1'] == '경기도']
```

```python
socar_zone_k_df = socar_zone_info_df[socar_zone_info_df['region1'] == '경기도']
```

```python
socar_zone_k_df.info()
```

```python
seoul_group_data = socar_zone_info_df.loc[socar_zone_info_df['region1'] == '경기도'].groupby('region2')['region2'].count()
seoul_group_data
```

```python
#2. 프로젝트(1)  # base nap : 서울시 중심부 위도, 경도
center = [37.541, 126.986]

map = folium.Map(
    location=center,
    zoom_start=10,
    )
print('슝~')
```

```python
#2. 프로젝트(1)  # MarkerCluster 객체 
marker_cluster = MarkerCluster().add_to(map)

#maker 객체 추가
for idx, row in socar_zone_k_df.iterrows():
    try:
        folium.Marker(
            location=[socar_zone_k_df.loc[idx, "lat"], socar_zone_k_df.loc[idx, "lng"]],
            popup=socar_zone_k_df.loc[idx, "zone_name"],
            icon=folium.Icon(color="cadetblue", icon="car", prefix="fa")            
        ).add_to(marker_cluster)
    except:
        pass

print('슝~')
```
