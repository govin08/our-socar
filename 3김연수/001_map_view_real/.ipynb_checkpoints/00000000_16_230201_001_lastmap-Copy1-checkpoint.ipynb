```python
#9) 시각화  ... 20-30 밀도 data 
k_pop_df = pd.read_csv("../002_data_dir/003_search_data/20_30_map_temp2.csv")
k_pop_df['adm_cd'] = k_pop_df['adm_cd'].astype(str)
k_pop_df.head()
```

```python
#9) 시각화  ... 20-30 밀도 시각화
folium.Choropleth(
    geo_data=k_geo_data,
    name="경기도 20-30대",
    data=k_pop_df,
    columns=["adm_cd", "pop15_39"],
    key_on="feature.properties.ADM_SECT_C",
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    line_color='white', 
    nan_fill_color='blue',
    line_weight=0,
    highlight=False, 
    smooth_factor=1.0,
    legend_name="인구수"
).add_to(map)
```

```python
#9) 시각화 ... 레이어 적용을 위한 그룹 만들기
# # 대표 그룹 만들기
fg = folium.FeatureGroup(name="groups")
map.add_child(fg)

# 쏘카 그룹 만들기
socar = plugins.FeatureGroupSubGroup(fg, "socar_zone")
map.add_child(socar)

# 행복주택 그룹 만들기
happy = plugins.FeatureGroupSubGroup(fg, "happy house")
map.add_child(happy)

# 대학교 그룹 만들기
university = plugins.FeatureGroupSubGroup(fg, "university")
map.add_child(university)

# 아파트 그룹 만들기
k_apt = plugins.FeatureGroupSubGroup(fg, "APT")
map.add_child(k_apt)

# 지하철 그룹 만들기
subway = plugins.FeatureGroupSubGroup(fg, "subway")
map.add_child(subway)

# 환승역 그룹 만들기 
transfer = plugins.FeatureGroupSubGroup(fg, "transfer")
map.add_child(transfer)

# 버스 그룹 만들기
bus = plugins.FeatureGroupSubGroup(fg, "bus")
map.add_child(bus)

# 주차장 그룹 만들기
parking = plugins.FeatureGroupSubGroup(fg, "parking")
map.add_child(parking)

# cluser
cls1 = plugins.FeatureGroupSubGroup(fg, "cluster1")
map.add_child(cls1)

# cluser
cls2 = plugins.FeatureGroupSubGroup(fg, "cluster2")
map.add_child(cls2)

# cluser
cls3 = plugins.FeatureGroupSubGroup(fg, "cluster3")
map.add_child(cls3)

# cluser
cls4 = plugins.FeatureGroupSubGroup(fg, "cluster4")
map.add_child(cls4)

# cluser
cls5 = plugins.FeatureGroupSubGroup(fg, "cluster5")
map.add_child(cls5)

# cluser
cls6 = plugins.FeatureGroupSubGroup(fg, "cluster6")
map.add_child(cls6)

# 레이어 콘트롤 박스 만들기
folium.LayerControl(collapsed=False).add_to(map)
```

```python
#9) 시각화 ...쏘카존 
mg1 = MarkerCluster()
for _, row in k_socar_zone_df.iterrows():
    mg1.add_child(    
        folium.Marker(location = [row['lat'], row['lng']],
               popup=row['zone_name'],
               icon=folium.Icon(color="cadetblue", icon="car", prefix="fa")
              )
    )

mg1.add_to(socar)
```

```python
#9) 시각화 ...행복주택 
mg2 = MarkerCluster()
for _, row in happy_house_df.iterrows():
    mg2.add_child(    
        folium.Marker(location = [row['lat'], row['lng']],
               popup=row['단지'],
               icon=folium.Icon(color='red',icon='home', prefix="fa")
              )
    )
    
mg2.add_to(happy)
```

```python
#9) 시각화 ...대학교
mg3 = MarkerCluster()
for _, row in k_university_df.iterrows():
    mg3.add_child(    
        folium.Marker(location = [row['lat'], row['lng']],
               popup=row['학교명'],
               icon=folium.Icon(color='green',icon='star',prefix="fa")
              )
    )

mg3.add_to(university)
```

```python
#9) 시각화 ...아파트
mg4 = MarkerCluster()
for _, row in k_apt_df.iterrows():
    mg4.add_child(    
        folium.Marker(location = [row['lat'], row['lng']],
               popup=row['공동주택명정보'],
               icon=folium.Icon(color='orange',icon='building',prefix="fa")
              )
    )

mg4.add_to(k_apt)
```

```python
#9) 시각화 ...지하철
mg5 = MarkerCluster()
for _, row in k_subway_info_df.iterrows():
    mg5.add_child(    
        folium.Marker(location = [row['lat'], row['lng']],
               popup=row['역명'],
               icon=folium.Icon(color='gray',icon='train',prefix="fa")
              )
    )

mg5.add_to(subway)
```

```python
#9) 시각화 ...버스
mg6 = MarkerCluster()
for _, row in bus_line_df.iterrows(): 
    mg6.add_child(    
        folium.Marker(location = [row['lat'], row['lng']],
               popup=row['정류장명'],
               icon=folium.Icon(color='lightgray',icon='van-shuttle',prefix="fa")
               
              )
    )

mg6.add_to(bus)
```

```python
#9) 시각화 ...환승역
mg7 = MarkerCluster()
for _, row in subway_transfer_df.iterrows():
    mg7.add_child(    
        folium.Marker(location = [row['lat'], row['lng']],
               popup=row['역명'],
               icon=folium.Icon(color='purple',icon='flag',prefix="fa")
              )
    )

mg7.add_to(transfer)
```

```python
#9) 시각화 : 주차장
mg8 = MarkerCluster()

# 주차장 이름, 위도, 경도를 가져와 maker 객체로 추가
for _, row in k_parking.iterrows():
    mg8.add_child(
        folium.Marker(
            location=[row["FCLTY_LA"], row["FCLTY_LO"]],
            popup=row["FCLTY_NM"],
            icon=folium.Icon(color="cadetblue", icon="ok")
        )
    )

mg8.add_to(parking)
```

```python
# Here we add cross-hatching (crossing lines) to display the Null values.
# nans = final_df[final_df["cluster"].isnull()]['ADM_SECT_C'].values
a_group = final_df[final_df["cluster"] == 0]['ADM_SECT_C'].values
b_group = final_df[final_df["cluster"] == 1]['ADM_SECT_C'].values
c_group = final_df[final_df["cluster"] == 2]['ADM_SECT_C'].values
d_group = final_df[final_df["cluster"] == 3]['ADM_SECT_C'].values
e_group = final_df[final_df["cluster"] == 4]['ADM_SECT_C'].values
f_group = final_df[final_df["cluster"] == 5]['ADM_SECT_C'].values
# gdf_nans = final_df[final_df['ADM_SECT_C'].isin(nans)]
gdf_a = final_df[final_df['ADM_SECT_C'].isin(a_group)]
gdf_b = final_df[final_df['ADM_SECT_C'].isin(b_group)]
gdf_c = final_df[final_df['ADM_SECT_C'].isin(c_group)]
gdf_d = final_df[final_df['ADM_SECT_C'].isin(d_group)]
gdf_e = final_df[final_df['ADM_SECT_C'].isin(e_group)]
gdf_f = final_df[final_df['ADM_SECT_C'].isin(f_group)]

sp = StripePattern(angle=45, color='grey', space_color='white')
sp.add_to(map)
# folium.features.GeoJson(name="Click for NaN values",data=gdf_nans, style_function=lambda x :{'fillPattern': sp},show=True).add_to(map2)
folium.features.GeoJson(name="Click for a",data=gdf_a, style_function=lambda x :{'fillPattern': sp},show=True).add_to(cls1)
folium.features.GeoJson(name="Click for b",data=gdf_b, style_function=lambda x :{'fillPattern': sp},show=True).add_to(cls2)
folium.features.GeoJson(name="Click for c",data=gdf_c, style_function=lambda x :{'fillPattern': sp},show=True).add_to(cls3)
folium.features.GeoJson(name="Click for d",data=gdf_d, style_function=lambda x :{'fillPattern': sp},show=True).add_to(cls4)
folium.features.GeoJson(name="Click for e",data=gdf_e, style_function=lambda x :{'fillPattern': sp},show=True).add_to(cls5)
folium.features.GeoJson(name="Click for f",data=gdf_f, style_function=lambda x :{'fillPattern': sp},show=True).add_to(cls6)

# We add a layer controller. 
# folium.LayerControl(collapsed=False).add_to(map)
# map2
```
